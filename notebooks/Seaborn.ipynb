## Exploring Seaborn Plots

### Histograms, KDE and densities

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], density=True, alpha=0.5)
```

```python
ax = plt.axes()
ax.set_ylim(0, 0.3)
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
ax = plt.axes()
ax.set_ylim(0, 0.3)
sns.distplot(data['x'])
sns.distplot(data['y']);
```

```python
sns.kdeplot(data);
```

```python
with sns.axes_style('white'):
    sns.jointplot('x', 'y', data, kind='kde')
```

```python
with sns.axes_style('white'):
    sns.jointplot('x', 'y', data, kind='hex')
```

### Pair plots

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue='species', size=2.5);
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

### Factor plots

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot("day", "total_bill", "sex", data=tips, kind="box")
    g.set_axis_labels("Day", "Total pay")
```

```python
with sns.axes_style(style='white'):
    sns.jointplot("total_bill", "tip", data=tips, kind='hex')
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg')
```

### Bar plots

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style(style='white'):
    g = sns.factorplot("year", data=planets, aspect=2, kind="count", color="steelblue")
    g.set_xticklabels(step=5)
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("year", data=planets, hue="method", aspect=4, kind="count", order=range(2001, 2015))
    g.set_ylabels("Number of Planets Discovered")
```

## Example: Exploring Marathon Finishing Times

```python
data = pd.read_csv('../data/marathon-data.csv')
data.head()
```

```python
data.dtypes
```

```python
import datetime
def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return datetime.timedelta(hours=h, minutes=m, seconds=s)
```

```python
data = pd.read_csv('../data/marathon-data.csv', converters={'split': convert_time, 'final': convert_time})
data.head()
```

```python
data.dtypes
```

```python
data['split_sec'] = data['split'].astype('int64') / 1E9
data['final_sec'] = data['final'].astype('int64') / 1E9
data.head()
```

```python
with sns.axes_style('white'):
    g = sns.jointplot("split_sec", "final_sec", data, kind='hex')
    g.ax_joint.plot(np.linspace(4000, 16000), np.linspace(8000, 32000), ':k')
```

```python
data['split_frac'] = 1 - 2 * data['split_sec'] / data['final_sec']
data.head()
```

```python
sns.distplot(data['split_frac'], kde=False)
plt.axvline(0, color="k", linestyle="--");
```

```python
sum(data.split_frac < 0)
```

```python
g = sns.PairGrid(data, vars=['age', 'split_sec', 'final_sec', 'split_frac'], hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

```python
sns.kdeplot(data.split_frac[data.gender == 'M'], label='men', shade=True)
sns.kdeplot(data.split_frac[data.gender == 'W'], label='women', shade=True)
plt.xlabel('split_frac');
```

```python
sns.violinplot("gender", "split_frac", data=data, palette=["lightblue", "lightpink"]);
```

```python
data['age_dec'] = data.age.map(lambda age: 10 * (age // 10))
data.head()
```

```python
men = data.gender == 'M'
women = data.gender == 'W'

with sns.axes_style(style=None):
    sns.violinplot("age_dec", "split_frac", hue="gender", data=data,
                   split=True, inner="quartile",
                   palette=["lightblue", "lightpink"]);
```

```python
g = sns.lmplot('final_sec', 'split_frac', col='gender', data=data,
               markers=".", scatter_kws=dict(color='c'))
g.map(plt.axhline, y=0.1, color="k", ls=":");
```
