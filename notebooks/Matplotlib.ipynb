#### MATLAB-style Interface

````python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1)  # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

### Adjusting the Plot: Line Colors ans Styles

````python
plt.plot(x, np.sin(x - 0), color='blue')           # specify color by name
plt.plot(x, np.sin(x - 1), color='g')              # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')           # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')        # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))  # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse');    # all HTML color names supported
````

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

````python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
````

## Visualizing Errors

### Basic Errorbars

````python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
````

````python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
````

````python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=5);
````

### Continuous Errors

````python
test = np.array([1, 3, 5, 6, 8])
print(test.shape)
print(test[np.newaxis, :])
print(test[:, np.newaxis])
````

````python
from sklearn.gaussian_process import GaussianProcess

model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1, random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE) # 2*sigma ~ 95% confidence region
````

````python
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit, color='gray', alpha=0.2)
plt.xlim(0, 10);
````

## Density and Contour Plots

````python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
````

### Visualizing a Three-Dimensional Function

````python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
````

````python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
````

````python
plt.contour(X, Y, Z, colors='black');
````

````python
a = np.array([1, 2, 3, 4, 5])
b = np.array([7, 8, 9])
A, B = np.meshgrid(a, b)
print(A)
print(B)
````

````python
plt.contour(X, Y, Z, 20, cmap='RdGy');
````

Colormaps: ```plt.cm```.TAB

````python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
````

````python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', 
           interpolation='gaussian')  # extent=[xmin, xmax, ymin, ymax]; interpolation for smoothing
plt.colorbar()
plt.axis(aspect='image');
````

````python
plt.figure(figsize=(7, 7))
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=10)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', interpolation='gaussian')
plt.colorbar();
````

````python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
````

### Two-Dimensional Histograms and Binnings

````python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
````

### ```plt.hist2d```

````python
print(x.shape)
print(y.shape)
````

````python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin');
````

````python
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
print(counts.shape)
print(counts[:3])
````

### ```plt.hexbin```

````python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cp = plt.colorbar(label='count in bin');
````

### Kernel density estimation

````python
from scipy.stats import gaussian_kde

data = np.vstack([x, y])
kde = gaussian_kde(data)

xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

plt.imshow(Z.reshape(Xgrid.shape),
           origin='lower', aspect='auto',
           extent=[-3.5, 3.5, -6, 6],
           cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
````
